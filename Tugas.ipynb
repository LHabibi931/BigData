```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import library

```python
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Panggil tabel Data.csv kemudian disimpan dalam variabel x dan y

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

mengubah word dalam bentuk matriks dengan cara mengimport column transformer dari library sklearn.compose kemudian mengimport onehotencoder dari library sklearn.preprocessing untuk encoder dari nominal ke bentuk matriks. Kemudian kita panggil kolom transformer, kita simpan di ct untuk encodernya menggunakan onehotencoder, dengan reminder passtrough. kemudian baru kita ubah format menjadi array dengan library numpy.

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

membagi data set menjadi 2, training dan testing dengan memanggil library sklearn di model selection kemudian import train test split dan lanjut dengan memasukkan masing masing ke variabel x dan y.
