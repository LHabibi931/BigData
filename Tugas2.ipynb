# Covert non-numerical values into numerical values

```python
def get_encoded_dict(values):
    return {val:i for i, val in enumerate(values)}
```

 get_encoded_dict() adalah sebuah fungsi menerima satu parameter bernama values, yang merupakan sebuah list yang akan di-encode menjadi sebuah dictionary. kemudian mengembalikan sebuah dictionary yang dihasilkan dengan menggunakan expression dictionary. Expression dictionary adalah cara sintaks Python untuk menuliskan sebuah dictionary dengan menggunakan syntax `{key:value for item in iterable}`.

```python
gender_dict = get_encoded_dict(df_data.Gender.unique())
married_dict = get_encoded_dict(df_data.Ever_Married.unique())
grad_dict = get_encoded_dict(df_data.Graduated.unique())
Profession_dict = get_encoded_dict(df_data.Profession.unique())
spend_dict = get_encoded_dict(df_data.Spending_Score.unique())
cat_dict = get_encoded_dict(df_data.Category.unique())
segment_dict = get_encoded_dict(df_data.Segmentation.unique())
segment_dict
```

kode tersebut membuat beberapa dictionary yang masing-masing menyimpan encoding untuk setiap unique value dari kolom-kolom tertentu dari objek DataFrame df_data.

Untuk membuat dictionary tersebut, kode tersebut memanggil fungsi get_encoded_dict() dengan menggunakan unique values dari masing-masing kolom sebagai parameter. Fungsi get_encoded_dict() kemudian akan mengembalikan dictionary yang meng-encode setiap elemen dari parameter tersebut menjadi integer dengan menggunakan indeks dari setiap elemen sebagai value dari entry dalam dictionary.

```python
encoded_dict = {
    'Gender' : gender_dict,
    'Ever_Married': married_dict,
    'Graduated' : grad_dict,
    'Profession': Profession_dict,
    'Spending_Score': spend_dict,
     'Category': cat_dict,
    'Segmentation': segment_dict
}
```

kode tersebut digunakan untuk membuat sebuah dictionary yang menyimpan encoding untuk setiap unique value dari beberapa kolom dari objek DataFrame df_data.

```python
df_data = df_data.replace(encoded_dict)
df_data.sample(10)
```

kode tersebut akan mengubah elemen-elemen dari objek df_data yang cocok dengan key dari encoded_dict menjadi value dari key tersebut

```python
df_data.columns
```

df_data.columns akan mengembalikan nama-nama kolom dari objek df_data dalam bentuk sebuah Index objek.

```python
feature_cols = ['Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession','Work_Experience', 'Spending_Score', 'Family_Size', 'Category']
target_col = ['Segmentation']
```

 feature_cols akan berisi nama-nama kolom yang akan digunakan sebagai fitur dalam suatu proses machine learning, sedangkan target_col akan berisi nama satu kolom yang akan digunakan sebagai target dalam proses tersebut.

```python
from sklearn.model_selection import train_test_split
data_x = df_data.loc[:, feature_cols]
data_y = df_data.loc[:, target_col]
train_x,test_x,train_y,test_y = train_test_split(data_x, data_y, test_size=0.2)
```

Fungsi train_test_split() adalah fungsi yang digunakan untuk membagi data menjadi data training dan data testing. Fungsi ini menerima empat parameter yaitu data_x, data_y, test_size, dan random_state.

    data_x adalah fitur dari data yang akan dibagi.
    data_y adalah target dari data yang akan dibagi.
    test_size adalah proporsi dari data yang akan dimasukkan ke dalam data testing. Nilai test_size harus berada di antara 0 dan 1.
    random_state adalah seed yang digunakan oleh fungsi untuk menentukan urutan acak dari data.

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

menampilkan tran_x,y dan test_x,y

```python
from sklearn.naive_bayes import CategoricalNB
cnb = CategoricalNB()
cnb.fit(train_x.to_numpy(), train_y.to_numpy().reshape(-1))
```

```python
actual_y = test_y.to_numpy().reshape(-1)
predicted_y = cnb.predict(test_x.to_numpy())
```

 actual_y diinisialisasi dengan mengubah objek DataFrame test_y menjadi array numpy dengan menggunakan method to_numpy(), kemudian mengubah bentuk array tersebut dengan menggunakan method reshape(-1). Method reshape(-1) akan mengubah bentuk array menjadi sebuah vektor dengan panjang sesuai dengan jumlah elemen dari array tersebut.

Sedangkan predicted_y diinisialisasi dengan mengubah objek DataFrame test_x menjadi array numpy dengan menggunakan method to_numpy(), kemudian memprediksi nilai output dari model machine learning yang telah dibuat pada data testing dengan menggunakan method predict() pada objek model tersebut dan menyimpan hasil prediksi tersebut dalam variabel predicted_y.

# Confusion matrix,Classification_Report

```python
from sklearn import metrics
metrics.accuracy_score(actual_y, predicted_y)
```

Jika hasil accuracy bernilai 0.54, artinya hasil prediksi dari model machine learning tersebut memiliki akurasi sebesar 54%. Hal ini berarti bahwa dari seluruh hasil prediksi yang dibuat oleh model machine learning tersebut, sebanyak 54% di antaranya sesuai dengan target dari data testing.

Nilai accuracy yang rendah menunjukkan bahwa model machine learning tersebut tidak terlalu baik dalam memprediksi target dari data testing. Hal ini bisa disebabkan karena beberapa faktor seperti kurangnya data yang tersedia, kurang baiknya fitur yang digunakan, atau model machine learning yang kurang tepat. Namun demikian, nilai accuracy tersebut masih bisa diperbaiki dengan cara menambah data yang tersedia, menggunakan fitur yang lebih baik, atau menggunakan model machine learning yang lebih tepat.
